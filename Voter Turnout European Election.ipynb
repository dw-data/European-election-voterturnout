```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

## Voter turnout among total voting age population for European elections

Voter turnout: European Parliament

URL:
https://election-results.eu/tools/download-datasheets/

```python
turnout = pd.read_csv('data/turnout-country.csv', delimiter=";")
```

```python
turnout
```

```python
fig, axes = plt.subplots(nrows=6, ncols=5, sharex=True, sharey=True, figsize=(15,15))
axes_list = [item for sublist in axes for item in sublist] 

for countryname, selection in turnout.groupby("COUNTRY_ID"):
    ax = axes_list.pop(0)
    selection.plot(x='YEAR', y='RATE', label=countryname, ax=ax, legend=False, clip_on=False)
    ax.set_title(countryname)
    ax.tick_params(
        which='both',
        bottom='off',
        left='off',
        right='off',
        top='off',
    )
    ax.grid(linewidth=0.5)
    ax.set_xlim((1979, 2020))
    ax.set_ylim((0,100))
    ax.set_xlabel("")
    ax.set_xticks(range(1979, 2020, 10))
    ax.set_yticks(range(20, 120, 20))
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

for ax in axes_list:
    ax.remove()

# plt.savefig("charts/drafts/Voter-Turnout-EU-Elections.png", transparent = True)
```

```python
latest2 = turnout[(turnout['YEAR']==2014)|(turnout['YEAR']==2019)]
latest2
```

```python
fig, axes = plt.subplots(nrows=7, ncols=4, sharex=True, sharey=True, figsize=(7,50))
axes_list = [item for sublist in axes for item in sublist] 

grouped = latest2.head(200).groupby("COUNTRY_ID")
ordered_country_names = grouped['RATE'].last().sort_values(ascending=False).index

for countryname in ordered_country_names:
    selection = grouped.get_group(countryname)
    ax = axes_list.pop(0)
    selection.plot(x='YEAR', y='RATE', label=countryname, ax=ax, legend=False, clip_on=False)
    ax.set_title(countryname)
    ax.tick_params(
        which='both',
        bottom='off',
        left='off',
        right='off',
        top='off',
    )
    ax.grid(linewidth=0.5)
    ax.set_xlim((2014, 2019))
    ax.set_ylim((0,100))
    ax.set_xlabel("")
    ax.set_xticks(range(2014,2019,5))
    ax.set_yticks(range(20, 120, 20))
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    max_year = selection['YEAR'].max()
    turnout_value = float(selection[selection['YEAR'] == max_year]['RATE'])
    ax.set_ylim((0, 100))
    ax.scatter(x=[max_year], y=[turnout_value], s=50, clip_on=False, linewidth=0)
    ax.annotate(str(int(turnout_value)) + "%", xy=[max_year, turnout_value], xytext=[5,0], textcoords='offset points')

    
for ax in axes_list:
    ax.remove()

plt.savefig("charts/drafts/Voter-Turnout-EU-Elections.pdf", transparent = True)
```
